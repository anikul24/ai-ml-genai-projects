```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA

# Load Iris dataset
iris = load_iris()
X = iris.data   # 4 features (sepal length, sepal width, petal length, petal width)
y = iris.target # 3 classes

# PCA: reduce to 2 components
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# LDA: reduce to 2 components (max = classes - 1 = 2 for 3 classes)
lda = LDA(n_components=2)
X_lda = lda.fit_transform(X, y)

# Plot PCA
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.scatter(X_pca[:,0], X_pca[:,1], c=y, cmap="viridis", edgecolor="k")
plt.title("PCA (unsupervised)")
plt.xlabel("PC1")
plt.ylabel("PC2")

# Plot LDA
plt.subplot(1,2,2)
plt.scatter(X_lda[:,0], X_lda[:,1], c=y, cmap="viridis", edgecolor="k")
plt.title("LDA (supervised)")
plt.xlabel("LD1")
plt.ylabel("LD2")

plt.show()
```
