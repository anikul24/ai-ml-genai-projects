# Data Preprocessing Tools

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt #charts
import pandas as pd

```

## Importing the dataset

```


dataset = pd.read_csv('/content/sample_data/Data.csv')
# dataset.head()


x= dataset.iloc[:,:-1].values #all columns exclusing last
y= dataset.iloc[:,-1].values # only last column


print(x)
print(y)
```

## Taking care of missing data

```
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(x[:,1:3])
x[:,1:3] = imputer.transform(x[:,1:3])
print(x)
```

## Encoding categorical data

### Encoding the Independent Variable

```
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough') ##to transform categorical data into a numerical format that machine learning algorithms can process
X =ct.fit_transform(x)



print(X)
print(type(X))
```

### Encoding the Dependent Variable

```
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)


print(y)
```

## Splitting the dataset into the Training set and Test set

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)

print(X_train)
```

## Feature Scaling

```
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

X_train[:,3:] = sc.fit_transform(X_train[:,3:]) # this will compute stand deviation and mean of age and salary
X_test[:,3:] = sc.transform(X_test[:,3:]) ## In above step fit does stand deviation and mean and trasnform scales it to all values,
##Above line trasnform is used becasue we already have trained model (SC) to stand deviation and mean  in line 4 or X_train

print(X_train)

```
