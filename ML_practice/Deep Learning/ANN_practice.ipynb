```python
## include libraraies

import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
tf.__version__
```

```python
dataset = pd.read_csv('/content/Churn_Modelling.csv')
X = dataset.iloc[:, 3:-1].values
y = dataset.iloc[:, -1].values

print("X Before change:", X)
```

```python
##label gender column


from sklearn.preprocessing import LabelEncoder


print("X before change:", X[:,2])
le = LabelEncoder()
X[:,2]=le.fit_transform(X[:,2])


le.inverse_transform(X[:,2].astype(int))

```

```python
##transform  geography(category) column

from sklearn.compose import ColumnTransformer

from sklearn.preprocessing import OneHotEncoder

ct = ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[1])], remainder='passthrough')



X = np.array(ct.fit_transform(X))

print(X)
```

```python
## test train split


from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)

```

```python
##feture scaling ( Always compulsory in deep learning)

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test= sc.transform(X_test)


sc.inverse_transform(X_train)
```

```python
##Build ANN

ann = tf.keras.models.Sequential()
```

```python
## add input and first hidden layer

ann.add(tf.keras.layers.Dense(units=6, activation='relu')) ##here unit is hyperparameter , this has to be tried and tuned , 6 is tried and tested

```

```python
## add second hidden layer

ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
```

```python
##Add output layer

ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid')) ## for non binary activation should be soft max
```

```python
##training ANN

## compile ANN

ann.compile(optimizer='adam', loss= 'binary_crossentropy', metrics=['accuracy'])  ## for non binary loss should be crossentropy

```

```python
##train the ANN on training set

ann.fit(X_train, y_train,batch_size=32, epochs = 100)
```

```python
##predict test result

y_pred = ann.predict(X_test)



y_pred = (y_pred > 0.5).astype(int)



print ("y_pred shape",y_pred.shape)
print ("y_test shape",y_test.shape)

# np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
